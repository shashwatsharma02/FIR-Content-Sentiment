```python
import pandas as pd
df = pd.read_csv('firContext.csv',sep='\t')
```

```python
df.head()
```

```python
#checking null val:
df.isnull().sum()
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
with open('firContext.csv') as f:
    doc = nlp(f.read())
```

```python
len(doc)
```

```python
#checking white space strings:
blanks=[]

for ind,review in df.itertuples():
    if type(review) == str:
        if review.isspace():
            blanks.append(ind)
    
```

```python
len(blanks)
```

```python
doc_sentences = [sent for sent in doc.sents]
```

```python
#Tokenisation:
```

```python
for sent in doc.sents:
    print(f"Sentence: {sent.text}")
    for token in sent:
        print(f"{token.text:{10}} {token.pos_:{10}} {token.dep_:{10}} {token.lemma_:{15}}")
    print("\n")
```

```python
#Removal of stop words:
```

```python
stop_words = spacy.lang.en.stop_words.STOP_WORDS
```

```python
for sent in doc_sentences:
    print(f"Sentence: {sent.text}")
    # Iterate through the tokens in each sentence
    for token in sent:
        # Check if the token is not a stop word
        if token.text.lower() not in stop_words:
            print(f"{token.text:{10}} {token.pos_:{5}} {token.dep_:{10}} {token.lemma_:{15}}")
    print("\n")
```
